```python
import json
with open("users.json", "r") as f:
    users = json.load(f)

print(users)
```

```python
# cleaning the data
def clean_data(users):
    cleaned = []

    for item in users:
        name = item["name"]
        age = item["age"]
        city = item["city"]

        if name and name.strip() and city and city.strip() and str(age).isdigit():
            cleaned.append({
                "name": name.strip(),
                "age": int(age),
                "city": city.strip()
            })

    return cleaned


print(clean_data(users))
```

```python
# Analysis of data
cleaned = clean_data(users)
#Total users
print("Total users: ", len(cleaned))
#Average age
total_age = 0
for item in cleaned:
    total_age += item["age"]
avg_age = total_age / len(cleaned)
print("Average age is : ", avg_age)
# City count
city_count = {}
for item in cleaned :
    city = item["city"]
    if city in city_count:
        city_count[city] += 1
    else:
        city_count[city] = 1
print("City count : " , city_count)
        
```

```python
with open("cleaned_users.json" , "w") as f:
    json.dump(cleaned , f , indent = 4)
```
