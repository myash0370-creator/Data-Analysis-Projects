## Data Collection

The data for this project was collected from the National Crime Records Bureau (NCRB), Government of India. 
The dataset contains yearly crime statistics for the state of Uttarakhand.

The collected data was then processed and transformed into a monthly format using Python for further analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load data
df = pd.read_csv("data.txt")

# Step 2: Months list
months = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

# Step 3: Create monthly data
rows = []

for _, row in df.iterrows():
    year = row["Year"]
    total = row["Total_Crimes"]
    
    for month in months:
        value = total / 12
        rows.append([year, month, row["State"], round(value)])

# Step 4: DataFrame
monthly_df = pd.DataFrame(rows, columns=["Year","Month","State","Total_Crimes"])

# 🔥 Step 5: Proper sorted output
monthly_df = monthly_df.sort_values(by=["Year","Month"])

# 🔥 Step 6: Full clean display
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

#print(monthly_df.to_string(index=False))
```

#Correct the data monthly order wise

```python
month_order = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]

monthly_df["Month"] = pd.Categorical(monthly_df["Month"], categories=month_order, ordered=True)

monthly_df = monthly_df.sort_values(by=["Year","Month"])
```

```python
# print(monthly_df.to_string(index=False))
```

```python
monthly_df.to_csv("final_monthly_data.csv", index=False)

print(" File saved successfully")
```

```python
yearly = monthly_df.groupby("Year")["Total_Crimes"].sum()
print(yearly)
```

# ==============================
# STATISTICAL ANALYSIS
# ==============================

```python


print(" BASIC STATISTICS")
print(monthly_df["Total_Crimes"].describe())


# Mean, Median, Mode
print("\n CENTRAL TENDENCY")
print("Mean:", monthly_df["Total_Crimes"].mean())
print("Median:", monthly_df["Total_Crimes"].median())
print("Mode:", monthly_df["Total_Crimes"].mode()[0])


# Standard Deviation & Variance
print("\n DISPERSION")
print("Standard Deviation:", monthly_df["Total_Crimes"].std())
print("Variance:", monthly_df["Total_Crimes"].var())


# Year-wise Statistics
print("\n YEAR-WISE STATS")
yearly_stats = monthly_df.groupby("Year")["Total_Crimes"].agg(["mean","std","min","max"])
print(yearly_stats)


# Growth Rate
print("\n GROWTH RATE (%)")
yearly_total = monthly_df.groupby("Year")["Total_Crimes"].sum()
growth = yearly_total.pct_change() * 100
print(growth)


# Skewness & Kurtosis
print("\n DISTRIBUTION SHAPE")
print("Skewness:", monthly_df["Total_Crimes"].skew())
print("Kurtosis:", monthly_df["Total_Crimes"].kurt())


# Moving Average
print("\n MOVING AVERAGE (Trend)")
# Year-wise total
yearly = monthly_df.groupby("Year")["Total_Crimes"].sum()
yearly_ma = yearly.rolling(window=3).mean()
print("📊 Year-wise Moving Average:")
print(yearly_ma)
```

```python
import matplotlib.pyplot as plt

# -----------------------------
# Year-wise Trend
# -----------------------------
yearly = monthly_df.groupby("Year")["Total_Crimes"].sum()

plt.figure()
plt.plot(yearly.index, yearly.values, marker='o')
plt.title("Year-wise Crime Trend")
plt.xlabel("Year")
plt.ylabel("Total Crimes")
plt.grid()
plt.show()


# -----------------------------
# Monthly Pattern
# -----------------------------
month_order = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
monthly_df["Month"] = pd.Categorical(monthly_df["Month"], categories=month_order, ordered=True)

monthly_pattern = monthly_df.groupby("Month")["Total_Crimes"].mean()

plt.figure()
plt.plot(monthly_pattern.index, monthly_pattern.values, marker='o')
plt.title("Monthly Crime Pattern")
plt.xlabel("Month")
plt.ylabel("Average Crimes")
plt.grid()
plt.show()


# -----------------------------
# Moving Average Graph
# -----------------------------
yearly_ma = yearly.rolling(window=3).mean()

plt.figure()
plt.plot(yearly.index, yearly.values, label="Actual")
plt.plot(yearly.index, yearly_ma, label="Moving Average")
plt.title("Crime Trend with Moving Average")
plt.xlabel("Year")
plt.ylabel("Total Crimes")
plt.legend()
plt.grid()
plt.show()


# -----------------------------
# Histogram
# -----------------------------
plt.figure()
plt.hist(yearly.values, bins=8)
plt.title("Crime Distribution")
plt.xlabel("Crime Count")
plt.ylabel("Frequency")
plt.show()
```
